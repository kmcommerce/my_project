## Evaluation Procedure

Composing an evaluation routine which can be used for all models in this project

```python
import sklearn.metrics as metrics

def evaluate_model(model, train_features, train_target, test_features, test_target):
    
    eval_stats = {}
    
    fig, axs = plt.subplots(1, 3, figsize=(20, 6)) 
    
    for type, features, target in (('train', train_features, train_target), ('test', test_features, test_target)):
        
        eval_stats[type] = {}
    
        pred_target = model.predict(features)
        pred_proba = model.predict_proba(features)[:, 1]
        
        # F1
        f1_thresholds = np.arange(0, 1.01, 0.05)
        f1_scores = [metrics.f1_score(target, pred_proba>=threshold) for threshold in f1_thresholds]
        
        # ROC
        fpr, tpr, roc_thresholds = metrics.roc_curve(target, pred_proba)
        roc_auc = metrics.roc_auc_score(target, pred_proba)    
        eval_stats[type]['ROC AUC'] = roc_auc

        # PRC
        precision, recall, pr_thresholds = metrics.precision_recall_curve(target, pred_proba)
        aps = metrics.average_precision_score(target, pred_proba)
        eval_stats[type]['APS'] = aps
        
        if type == 'train':
            color = 'blue'
        else:
            color = 'green'

        # F1 Score
        ax = axs[0]
        max_f1_score_idx = np.argmax(f1_scores)
        ax.plot(f1_thresholds, f1_scores, color=color, label=f'{type}, max={f1_scores[max_f1_score_idx]:.2f} @ {f1_thresholds[max_f1_score_idx]:.2f}')
        # setting crosses for some thresholds
        for threshold in (0.2, 0.4, 0.5, 0.6, 0.8):
            closest_value_idx = np.argmin(np.abs(f1_thresholds-threshold))
            marker_color = 'orange' if threshold != 0.5 else 'red'
            ax.plot(f1_thresholds[closest_value_idx], f1_scores[closest_value_idx], color=marker_color, marker='X', markersize=7)
        ax.set_xlim([-0.02, 1.02])    
        ax.set_ylim([-0.02, 1.02])
        ax.set_xlabel('threshold')
        ax.set_ylabel('F1')
        ax.legend(loc='lower center')
        ax.set_title(f'F1 Score') 

        # ROC
        ax = axs[1]    
        ax.plot(fpr, tpr, color=color, label=f'{type}, ROC AUC={roc_auc:.2f}')
        # setting crosses for some thresholds
        for threshold in (0.2, 0.4, 0.5, 0.6, 0.8):
            closest_value_idx = np.argmin(np.abs(roc_thresholds-threshold))
            marker_color = 'orange' if threshold != 0.5 else 'red'            
            ax.plot(fpr[closest_value_idx], tpr[closest_value_idx], color=marker_color, marker='X', markersize=7)
        ax.plot([0, 1], [0, 1], color='grey', linestyle='--')
        ax.set_xlim([-0.02, 1.02])    
        ax.set_ylim([-0.02, 1.02])
        ax.set_xlabel('FPR')
        ax.set_ylabel('TPR')
        ax.legend(loc='lower center')        
        ax.set_title(f'ROC Curve')
        
        # PRC
        ax = axs[2]
        ax.plot(recall, precision, color=color, label=f'{type}, AP={aps:.2f}')
        # setting crosses for some thresholds
        for threshold in (0.2, 0.4, 0.5, 0.6, 0.8):
            closest_value_idx = np.argmin(np.abs(pr_thresholds-threshold))
            marker_color = 'orange' if threshold != 0.5 else 'red'
            ax.plot(recall[closest_value_idx], precision[closest_value_idx], color=marker_color, marker='X', markersize=7)
        ax.set_xlim([-0.02, 1.02])    
        ax.set_ylim([-0.02, 1.02])
        ax.set_xlabel('recall')
        ax.set_ylabel('precision')
        ax.legend(loc='lower center')
        ax.set_title(f'PRC')        

        eval_stats[type]['Accuracy'] = metrics.accuracy_score(target, pred_target)
        eval_stats[type]['F1'] = metrics.f1_score(target, pred_target)
    
    df_eval_stats = pd.DataFrame(eval_stats)
    df_eval_stats = df_eval_stats.round(2)
    df_eval_stats = df_eval_stats.reindex(index=('Accuracy', 'F1', 'APS', 'ROC AUC'))
    
    print(df_eval_stats)
    
    return df_eval_stats
```

## Normalization

We assume all models below accepts texts in lowercase and without any digits, punctuations marks etc.

```python
# initialization
def initialize_nltk():
    try:
        nltk.data.find('tokenizers/punkt')
        nltk.data.find('corpora/stopwords')
        nltk.data.find('corpora/wordnet')
    except LookupError:
        #import os
        try:
            # Try default download first
            nltk.download(['punkt', 'stopwords', 'wordnet'], quiet=True)
        except PermissionError:
            # Fallback to current directory if no permissions
            nltk.download(['punkt', 'stopwords', 'wordnet'], 
                         download_dir=os.getcwd(), 
                         quiet=True)
            nltk.data.path.append(os.getcwd())

# Initialize NLTK silently
initialize_nltk()
```

```python
def normalize_text(text):
    # Initialize NLTK resources (safe to call multiple times)
    initialize_nltk()
    
    # Your normalization steps
    text = text.lower()
    tokens = word_tokenize(text)
    tokens = [word for word in tokens if word not in stopwords.words('english')]
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(word) for word in tokens]
    return ' '.join(tokens)

#df_reviews['review_norm'] = df_reviews['review'].apply(normalize_text)
start = time.time()
df_reviews['review_norm'] = df_reviews['review'].progress_apply(normalize_text)
#print("Normalization complete in", time.time() - start, "seconds", flush=True)
```

## My Reviews

```python
from tqdm import tqdm

# Load spaCy model
nlp = spacy.load("en_core_web_sm")

# Sample DataFrame
my_reviews = pd.DataFrame([
    'I did not simply like it. Endless punches, no real knockout.',
    'Well, I was bored and dozed off during the second round. The pacing is just dull.',
    'I was really fascinated with the technique ‚Äî the accuracy of each punch felt spot on',    
    'Even the fighters looked stiff and robotic, and the animations felt cheap. What a low quality product and cash grab.',
    'I didn\'t expect the reboot to be so good! The organizer clearly cared about boxing history and got it right',
    'The fight had some moments, but overall it felt repetitive and shallow. Not sure I‚Äôd put any interest on it again.',
    'What a lousy attempt at a boxing sim. Controls are clunky, commentary is unbearable, and the opponent technique is completely broken',
    'Releasing this on console was a bold move. Flat, uninspired rounds.'
], columns=['review'])

# function to retain negations
def lemmatize_and_clean(text):
    doc = nlp(text.lower())
    cleaned_tokens = []
    for token in doc:
        # some negation words to keep 
        if token.dep_ == "neg" or token.text in ["no", "not", "never", "didn't", "don't", "doesn't"]:
            cleaned_tokens.append(token.text) 
        
        elif token.is_alpha and not token.is_stop:
            cleaned_tokens.append(token.lemma_)
    return ' '.join(cleaned_tokens)


tqdm.pandas()
my_reviews['cleaned'] = my_reviews['review'].progress_apply(lemmatize_and_clean)


print(my_reviews[['review', 'cleaned']])
my_reviews
```

### Model 2

```python
texts = my_reviews['cleaned']
rfidf_clean = tfidf_vectorizer_2.transform(texts)
my_reviews['sentiment_prob'] = model_2.predict_proba(rfidf_clean)[:, 1]

for i, review in enumerate(texts.str.slice(0, 100)):
    print(f'{my_reviews["sentiment_prob"][i]:.2f}:  {review}')
```
